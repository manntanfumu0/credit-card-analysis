```python
import pandas as pd
df = pd.read_csv("../data/raw/Card_Aprov_Pre.csv")
```

```python
df.head()
```

```python
import unicodedata

def limpar_coluna(col):
    col = ''.join(
        c for c in unicodedata.normalize('NFKD', col)
        if not unicodedata.combining(c)
    )
    return (
        col.strip()
        .lower()
        .replace(" ", "_")
        .replace("(", "")
        .replace(")", "")
    )

df.columns = [limpar_coluna(col) for col in df.columns]
```

```python
df.head()
```

```python
colunas_binarias = ["aprovado", "empregadoa", "cliente_do_banco"]

for col in colunas_binarias:
    if df[col].dtype == "object":
        df[col] = df[col].map({
            "Sim": 1, "Não": 0,
            "Yes": 1, "No": 0
        })
```

```python
df.to_csv("../data/processed/dados_limpos.csv", index=False)
```
